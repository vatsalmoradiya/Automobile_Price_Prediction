```python
mean_cols = ['normalized-losses','stroke','horsepower','price']
median_cols = ['bore','peak-rpm','num-of-doors']
for col in mean_cols:
    df[col] = pd.to_numeric(df[col])
    df[col].fillna(df[col].mean(),inplace=True)
for col in median_cols:
    df[col] = pd.to_numeric(df[col])
    df[col].fillna(df[col].median(),inplace=True)
```

There are a total of 8 categorical features in this dataset. Few of them could be converted into integer values using
LabelEncoder(). Others requrie in depth analysis and absed on that analysis, dummy columns might be used.

The 8 features are: 
make, fuel-type, aspiration, body-style, drive-wheels, engine-location, engine-type, num-of-cylinders, fuel-system

```python
df['num-of-cylinders'].value_counts()
```

```python
from word2number import w2n
df['num-of-cylinders'] = df['num-of-cylinders'].apply(w2n.word_to_num)
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df[['fuel-type','aspiration','engine-location']] = df[['fuel-type','aspiration','engine-location']].apply(lb.fit_transform)
df['OHC_Code'] = np.where(df["engine-type"].str.contains("ohc"), 1, 0)
df.drop(['engine-type'],axis=1,inplace=True)
```

### Now only four of the features are remaining to be encoded. Two methods are possible:
1. Do feature selection using correlation and include these features with them then apply pipeline and linear regression.
2. Do one hot encoding by yourself then do feature selection using correlation and then apply linear regression.

### Method 1

```python
colormap = plt.cm.viridis
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
sorted_index = abs(df.corr()).sort_values('price',ascending=False).index
sorted_index
```

```python
features = []
for i in range(len(sorted_index)):
#     print(df.corr().loc['price',sorted_index[i]])
    if abs(df.corr().loc['price',sorted_index[i]]) > 0.2:
        select = True
        for j in range(i):
            if abs(df.corr().loc[sorted_index[i],sorted_index[j]])>0.9:
                select=False
        if select==True:
            features.append(sorted_index[i])
```

```python
features
```

### Selected features are:
'engine-size', 'curb-weight', 'horsepower', 'width', 'highway-mpg', 'num-of-cylinders', 'length', 'wheel-base', 'bore', 'engine-location'

### Adding the remaining 4 to be encoded in selected features: 
make, body-style, drive-wheels, fuel-system

```python
features_df_1 = df[features[1:] + ['body-style','drive-wheels','fuel-system'] + [features[0]]]
```

```python
features_df_1.head()
```

```python
## Cretaing Pipeline and applying Linear Regression
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

```python
X = features_df_1.drop(['price'],axis=1)
y = features_df_1['price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
column_trans = make_column_transformer((OneHotEncoder(handle_unknown='ignore'),
                                        ['body-style','drive-wheels','fuel-system']),
                                      remainder='passthrough')
```

```python
linreg = LinearRegression()
pipe = make_pipeline(column_trans, linreg)
```

```python
cross_val_score(pipe, X, y, cv=10, scoring='neg_mean_absolute_error').mean().round(2)
```

### Method 2

```python
df2 = df.copy(deep=True)
```

```python
oe_style = OneHotEncoder()
rem_feat = ['make','body-style','drive-wheels','fuel-system']
for feat in rem_feat:
    oe_results = oe_style.fit_transform(df2[[feat]])
    df2 = pd.concat([df2.drop([feat],axis=1), pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_[0])],axis=1)
```

```python
colormap = plt.cm.viridis
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df2.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white')
```

```python
sorted_index2 = abs(df2.corr().sort_values('price',ascending=False)).index
```

```python
features2 = []
for i in range(len(sorted_index2)):
    if abs(df2.corr().loc['price',sorted_index2[i]]) > 0.1:
        select = True
        for j in range(i):
            if abs(df2.corr().loc[sorted_index2[i],sorted_index2[j]]) > 0.9:
                select = False
        if select==True:
            features2.append(sorted_index2[i])
```

```python
features2.remove('price')
features2.append('price')
```

```python
features_df_2 = df2[features2]
```

```python
X_2 = features_df_2.drop(['price'],axis=1)
y_2 = features_df_2['price']
```

```python
X_2.head()
```

```python
y_2.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_2,y_2,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_2_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import math
mse = mean_squared_error(y_test,y_2_pred)
math.sqrt(mse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_2_pred)
round(score * 100 ,2)
```
